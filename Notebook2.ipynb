```python
from ImageBlocks import split_image, blocks_similarity
from Segmentation import change_shade, random_color, image_segmentation, similarity_with_neighbours, form_images
from Segmentation import overlapping_output
from pathlib import Path
import cv2
import numpy as np
```

```python
path = Path.cwd()/"dummy.jpg"
img_blocks, img = split_image(str(path), block_size=(10, 10), gray=False, draw_img=True)
```

```python
i = form_images(blocks=img_blocks)
cv2.imshow("Image", i)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_blocks.shape
```

```python
s = image_segmentation(img_blocks, color_shading=True, thresh=0.9)
```

```python
for row in s:
    for block in row:
        print(block, block.color, end=" ")
    print()
```

```python
for row in range(len(s)):
    for col in range(len(s[row])):
        s[row][col].print_color()
```

```python
output = form_images(s, change_color=True)
```

```python
cv2.imshow("Output", output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
